# Simple Algebra Operations

### Notes:
Python mein basic mathematical calculation bohot aasan hai. Hum direct arithmetic operators use karte hain:
- `+` (Addition): Numbers ko jama karne ke liye.
- `-` (Subtraction): Ek number se dusre ko nikalne ke liye.
- `*` (Multiplication): Multiply karne ke liye.
- `/` (Division): Division, jo humesha float (decimal value) return karta hai.
- `//` (Floor Division): Division ke baad decimal portion ko remove kar deta hai aur solid integer deta hai.
- `**` (Power/Exponentiation): Kisi number ki power nikalne ke liye.
- `%` (Modulus): Divide karne ke baad bacha hua remainder return karta hai.

**Real-Life Example:**
Aap ek grocery store par hain. 
- 5 items khareede aur unka total sum nikalna **Addition** hai (`+`).
- Agar total bill Rs. 600 bana aur aapne Rs. 1000 ka note diya, to bacha hua change nikalna **Subtraction** hai (`-`).
- Agar 3 packets biscuits khareede jinme har ek Rs. 50 ka hai, to total calculate karna **Multiplication** hai (`*`).
- Ek pizza ko 5 dosto mein barabar baantna aur har ek ka hissa nikalna **Division** hai (`/`).
- Kisi bache hue pizza slice (remainder) ko check karna **Modulus** hai (`%`).


```python
print("Hello World!")
a = complex(5,6)
print("Type of a is: ",type(a))
print("Addition of two numbers: ",6+5)
print("Subtraction of two numbers: ",6-5)
print("Multiplication of two numbers: ",6*5)
print("Division of two numbers: ",6/5)
print("Division when we want to remove digit after decimal point: ",6//5)
print("Power of any number is: ",2**3)
print("Remainder of numbers: ",6%5)
```

# Match Case Statement

### Notes:
Match case statement Python 3.10+ mein introduce kiya gaya tha, jo dusri languages ke `switch-case` ki tarah kaam karta hai. Ye codes ko zyada readable banata hai jab hume ek hi variable ke bohot saare options check karne hon. Isme `_` default case ko represent karta hai.

**Real-Life Example:**
Customer care IVR menu:
- Press 1 for English.
- Press 2 for Urdu/Hindi.
- Press 3 to speak with an agent.
- `_` (Default): Invalid option entered, please try again.


```python
x = 10
match x:
    case 0:
        print("x is zero.")
    case 4:
        print("X is 4.")
    case _ if x!= 90:
        print(x,'is not 90.')
    case _ if x!= 80:
        print(x,' is not 80.')
    case _:
        print(x)
```

# Function Arguements

### Notes:
Python mein arguments pass karne ke alag-alag tarike hote hain:
1. **Required Arguments**: Jinhe call karte waqt pass karna lazmi hota hai.
2. **Default Arguments**: Agar value pass na ho, to default value use hoti hai.
3. **Keyword Arguments**: Arguments ko unke parameter name ke saath pass kiya jata hai.
4. **Variable-length (*args, **kwargs)**: Multiple values ko tuple ya dictionary ke form mein accept kiya jata hai.

**Real-Life Example:**
Ek online pizza order form:
- Pizza Size: Required (aapko batana hi padega).
- Extra Cheese: Default value is `False` (agar aap kuch na kahein, to simple pizza milega).


```python
# 1. Required Arguements
def average(f,g):
    print("Average of numbers is: ",(f+g)/2)
average(5,6)

# Default Arguements
def average2(a=5,b=7):
    print("Average of two numbers: ",(a+b)/2)
average2()
average2(4,6)
average2(3)
average2(b=9)
def name(fname,sname='Hassan',tname='Ahsan'):
    print("Hello",fname,sname,tname)
name("Mohsan")

#keyword arguement
def avg(a=5,b=10):
    print("Average is: ",(a+b)/2)
avg(b=10,a=20)

# tupple as a arguement of function
def avg2(*number):
    sum = 0
    print(type(number))
    for i in number:
        sum = sum + i
    print("Average of numbers: ",sum/len(number))
avg2(3,45,6,7,32,89)

#for dictionary write as def average(**numbers):
```

# Recursion

### Notes:
Recursion ek aisi technique hai jahan ek function apne aap ko hi baar-baar call karta hai jab tak wo ek specific 'base case' (stop condition) tak na pohonch jaye.

**Real-Life Example:**
Matryoshka dolls (Russian nesting dolls). Ek badi doll ke andar choti doll, uske andar aur choti doll, jab tak sabse aakhri doll na aa jaye.


```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)
factorial(5)
def fibo(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fibo(n-1) + fibo(n-2)
n = 5
for i in range(n):
    print(fibo(i))
```

# Map

# Setters and Getters

### Notes:
Getters (`@property`) aur Setters (`@value.setter`) variables ki values ko safely access aur modify karne ke liye use hote hain, taaki security maintain rahe.

**Real-Life Example:**
ATM pin update karna. Pehle system check karta hai ki kya enter kiya gaya PIN valid criteria (4-digits) par poora utarta hai ya nahi.


```python
class MyClass:
    def __init__(self,value):
        self._value = value
    def show(self):
        print(f"Value is {self._value}")
    @property
    def ten_value(self):
        return 10 * self._value
    @ten_value.setter
    def ten_value(self,new_value):
        self._value = new_value/10
obj = MyClass(10)
obj.ten_value = 67
print(obj.ten_value)
obj.show()
```

# Multi-Threading Module

### Notes:
Multi-threading ka matlab hai multiple tasks ko concurrently chalana. Ye I/O bound operations (jaise files download karna, network call karna) ke liye best hai.

**Real-Life Example:**
Web browser mein ek tab par song sunna aur doosre tab par file download hona dono concurrent ho rahe hain.


```python
import threading 
import time
from concurrent.futures import ThreadPoolExecutor
def func(seconds):
    print(f"Sleeping for {seconds} seconds")
    time.sleep(seconds)
    return seconds
def main():
    time1 = time.perf_counter()

    # #Normal Code
    func(4)
    func(1)
    func(2)
    time2 = time.perf_counter()
    print(time2 - time1)

    #Same code with threads
    t1 = threading.Thread(target=func,args=[4])
    t2 = threading.Thread(target=func,args=[2])
    t3 = threading.Thread(target=func,args=[1])
    t1.start()
    t2.start()
    t3.start()
    t1.join()
    t2.join()
    t3.join()
    time2 = time.perf_counter()
    print(time2 - time1)
    
def poollingDemo():
    with ThreadPoolExecutor() as executor:
        future1 = executor.submit(func,3)
        future2 = executor.submit(func,2)
        future3 = executor.submit(func,4)
        print(future1.result())
        print(future2.result())
        print(future3.result())
        l = [3,5,1,2]
        results = executor.map(func,l)
        for result in results:
            print(result)
poollingDemo()
```

# Multi-Processing

### Notes:
Multi-processing multiple processes ko alag cores par parallelly execute karta hai. Ye CPU-bound heavy calculations ke liye best hai.

**Real-Life Example:**
Video rendering ya heavy machine learning training, jahan computing cores ka load speed badhane ke liye distribute kiya jata hai.


```python
import multiprocessing
import requests
from concurrent.futures import ProcessPoolExecutor
def download_file(url,name):
    print(f"Download {name} Started...")
    response = requests.get(url)
    open(f"{name}.jpg",'wb').write(response.content)
    print(f"Finished Downloading {name}")
url = "https://picsum.photos/200/300"

# Method 1 to apply Multi Processing 
pros = []
for i in range(5):
    download_file(url,i)
    p = multiprocessing.Process(target=download_file,args=[url,i])
    p.start()
    pros.append(p)
for p in pros:
    p.join()

# Method 2 to apply Multi Processing
print("\n--- Starting ProcessPoolExecutor ---")
with ProcessPoolExecutor() as executor:
    l1 = [url for _ in range(10)]
    l2 = [f"pool_{i}" for i in range(10)]
    results = executor.map(download_file, l1, l2)
    for r in results:
        print(r)
```
